```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from emocoder.src import utils, constants
from emocoder.experiments import constants as xconstants
```

```python
split = "test"  # dev or test
```

```python
baseline = pd.read_csv(xconstants.MAPPING_BASELINE_BASEPATH / split / "overview.csv", index_col=0)
baseline.rename(inplace=True, index={"anew_stevenson_baseline_vad_to_be5": "anew-stevenson",
                                     "anew_stevenson_baseline_va_to_be5": "anewVA-stevenson",
                                     "anew_stevenson_baseline_be5_to_vad": "stevenson-anew",
                                     "anew_stevenson_baseline_be5_to_va": "stevenson-anewVA",
                                     "FER-VAD-BE": "ferVAD-ferBE",
                                     "FER-BE-VAD": "ferBE-ferVAD",
                                     "AffectNet-VA-BE": "affectnetVA-affectnetBE",
                                     "AffectNet-BE-VA": "affectnetBE-affectnetVA"})
baseline["mean"] = baseline.mean(axis=1)
baseline
```

```python
if split == "dev":
    multitask = pd.read_csv(xconstants.MAPPING_MULTITASK_BASEPATH / split/ "overview.csv", index_col=[1])
    multitask = multitask.drop(columns=["Unnamed: 0", "mean"]).drop(index="overall_mean")
    multitask = multitask.rename(index={"ANEW_Stevenson_vad_be5": "anew-stevenson",
                            "ANEW_Stevenson_be5_vad": "stevenson-anew",
                            "ANEW_VA_Stevenson_va_be5": "anewVA-stevenson",
                            "ANEW_VA_Stevenson_be5_va": "stevenson-anewVA",
                            "FER_BE_VAD_vad_be_fer13": "ferVAD-ferBE",
                            "FER_BE_VAD_be_fer13_vad": "ferBE-ferVAD",
                            "AffectNet_Mapping_va_be_affectnet": "affectnetVA-affectnetBE",
                            "AffectNet_Mapping_be_affectnet_va": "affectnetBE-affectnetVA"})
    multitask["mean"] = multitask.mean(axis=1)
if split == "test":
    multitask = pd.read_csv(xconstants.MAPPING_MULTITASK_BASEPATH / split/ "overview.csv", index_col=[0])
    multitask = multitask.rename(index={"ANEW_Stevenson_vad_be5": "anew-stevenson",
                            "ANEW_Stevenson_be5_vad": "stevenson-anew",
                            "ANEW_VA_Stevenson_va_be5": "anewVA-stevenson",
                            "ANEW_VA_Stevenson_be5_va": "stevenson-anewVA",
                            "FER_BE_VAD_vad_be_fer13": "ferVAD-ferBE",
                            "FER_BE_VAD_be_fer13_vad": "ferBE-ferVAD",
                            "AffectNet_Mapping_va_be_affectnet": "affectnetVA-affectnetBE",
                            "AffectNet_Mapping_be_affectnet_va": "affectnetBE-affectnetVA"})
    multitask["mean"] = multitask.mean(axis=1)

multitask
```

```python
overall = pd.DataFrame(index=multitask.index, data={"baseline": baseline.loc[multitask.index, "mean"],
                                                    "multitask": multitask["mean"]})
overall.loc["mean"] = overall.mean(axis=0)
overall.loc["median"] = overall.iloc[:-1].median(axis=0)
overall = overall.round(2)
overall
```
